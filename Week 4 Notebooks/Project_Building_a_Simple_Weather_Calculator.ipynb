# Project Overview

Your programme should allow the user to choose between two weather-related calculations:

1. **Temperature Conversion:**  
   Convert a temperature from Fahrenheit to Celsius.  
   **Formula:**  
   Celsius = (Fahrenheit - 32) x 5/9
   
   *Explanation:* Subtract 32 from the Fahrenheit temperature, multiply by 5, then divide by 9 to obtain the equivalent Celsius temperature.

2. **Wind Chill Calculation:**  
   Estimate the wind chill using a simplified version of the formula.  
   **Formula:**  
   
   Wind Chill = 35.74 + 0.6215 x T - 35.75 x V^{0.16} + 0.4275 x T x V^{0.16}
   
   where:  
   - T is the temperature in Fahrenheit  
   - V is the wind speed in mph  
   
   *Explanation:* This formula provides an estimate of how cold it feels when wind speed is taken into account along with the actual air temperature.

```python
# --- 1. Math Functions ---

def fahrenheit_to_celsius(f_temp):
    """Converts Fahrenheit to Celsius."""
    celsius = (f_temp - 32) * (5 / 9)
    return celsius

def calculate_wind_chill(t, v):
    """Calculates wind chill based on Temperature (F) and Wind Speed (mph)."""
    # Formula: 35.74 + 0.6215T - 35.75(V^0.16) + 0.4275T(V^0.16)
    wind_chill = 35.74 + (0.6215 * t) - (35.75 * (v ** 0.16)) + (0.4275 * t * (v ** 0.16))
    return wind_chill

# --- 2. Menu and Input Functions ---

def display_menu():
    """Displays the menu and gets the user's choice."""
    print("\n--- Weather Calculator ---")
    print("1. Convert Temperature (Fahrenheit to Celsius)")
    print("2. Calculate Wind Chill")
    print("3. Exit")
    choice = input("Enter your choice (1-3): ")
    return choice

def get_number(prompt):
    """Safely gets a number from the user."""
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Invalid input. Please enter a valid number.")

# --- 3. Main Program Logic ---

while True:
    user_choice = display_menu()

    if user_choice == "1":
        # Temperature Conversion
        f_temp = get_number("Enter the temperature in Fahrenheit: ")
        c_temp = fahrenheit_to_celsius(f_temp)
        # The :.2f formats the result to 2 decimal places so it looks clean!
        print(f"\nResult: {f_temp}°F is equivalent to {c_temp:.2f}°C.")

    elif user_choice == "2":
        # Wind Chill
        t = get_number("Enter the temperature in Fahrenheit: ")
        v = get_number("Enter the wind speed in mph: ")
        chill = calculate_wind_chill(t, v)
        print(f"\nResult: The estimated wind chill is {chill:.2f}°F.")

    elif user_choice == "3":
        print("\nExiting the Weather Calculator. Have a great day!")
        break # This breaks the while loop and ends the program

    else:
        print("\nInvalid choice. Please select 1, 2, or 3.")
```
